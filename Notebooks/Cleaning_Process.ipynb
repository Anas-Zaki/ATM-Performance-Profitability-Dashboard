---
We are identifying and dropping columns that have fewer than `3 unique values` because such columns provide very little or no variation and do not contribute meaningful insights during analysis or modeling.

By removing them, we:

- Reduce unnecessary noise and simplify the dataset
- Improve visualization clarity and model performance
- Keep only columns that can provide useful patterns or differences

We exclude `Year` and `ATM AMC` because, even if they have limited unique values, they are still important features for analysis.

```python
unique_counts = df.nunique()

# Find columns with fewer than 3 unique values
few_unique_cols = unique_counts[unique_counts < 3].index.tolist()

# Remove 'year' and 'ATM AMC' from that list (case-insensitive)
few_unique_cols = [col for col in few_unique_cols if col.lower() != 'year' and col != 'ATM AMC']

# Drop those columns
df = df.drop(columns=few_unique_cols)
```

```python
df.isnull().sum()
```

---
Our dataset has `11,076` total rows, but the `Category` column has only `10,703` non-null values, meaning some rows have missing category information.

We remove these rows because:

- The `Category` field is important for classification and analysis.
- Rows without a category provide incomplete information and can distort insights.
- Keeping only complete records ensures cleaner, more reliable analysis and accurate visualization results.

```python
df = df.drop(columns='Category')
```

---
Some column names have extra spaces or irregular spacing, which can cause errors when referencing columns in code.

By splitting and rejoining the column names, we remove unnecessary spaces and ensure clean, consistent, and easy-to-use column names for analysis and visualization.

```python
col = df.columns.str.split()
col = [" ".join(i) for i in col]
df.columns = col
```

```python
df.info()
```

Columns from `Fin Txn` to `Penalty` contain numeric data stored as strings.

Converting them to proper numeric types (int or float) ensures:
- Mathematical operations like sum, mean, or correlation can be performed correctly.
- Visualizations (charts, plots) are accurate.
- The dataset is clean and ready for analysis and further EDA.

```python
num_cols = col[4:29]
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.dtypes.head(40)
```

---
The Month column contains only four unique dates, so instead of full date values, we only need the month information.

Extracting the month:
- Simplifies the dataset for trend analysis.
- Makes grouping by month easier for visualization and insights.
- Avoids unnecessary detail since only four distinct dates exist.

```python
df["Month"] = df["Month"].astype(str)
df["Month"] = df["Month"].str.replace(r'^\d{4}', '2024', regex=True)
df["Month"] = pd.to_datetime(df["Month"]).dt.month_name()
```

```python
df.to_csv("ATM_Data_Cleaned.csv", index=False)
```
